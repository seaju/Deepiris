```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs

data,labels=make_blobs(n_samples=10000,n_features=2,centers=2,random_state=0)
plt.scatter(data[:,0],data[:,1],c=labels,label=labels)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data, labels, test_size=0.2)
```

```python
X_val=X_train[:1000]
X_train=X_train[1000:]
y_val=y_train[:1000]
y_train=y_train[1000:]
```

```python
from keras import models
from keras import layers
model=models.Sequential()
model.add(layers.Dense(1,input_dim=2,activation='sigmoid'))
```

```python
model.compile(optimizer='SGD',loss='binary_crossentropy',metrics=['accuracy'])
model.fit(X_train, y_train, epochs=40, batch_size=100)
```

```python
history=model.fit(X_train,y_train,epochs=10,batch_size=100,validation_data=(X_val,y_val))
```

```python
result=model.evaluate(X_test,y_test)
result
```

```python
import matplotlib.pyplot as plt
history_dict=history.history
loss=history_dict['loss']
val_loss=history_dict['val_loss']
epochs=range(1,len(loss)+1)
plt.plot(epochs,loss,'bo',label='Training loss')
plt.plot(epochs,val_loss,'r',label='validation loss')
plt.title('Training and validation loss')
plt.axis([0,12,0.0,3.0])
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
#
plt.clf()
acc=history_dict['acc']
val_acc=history_dict['val_acc']
plt.plot(epochs,acc,'bo',label='Training acc')
plt.plot(epochs,val_acc,'r',label='Validation acc')
plt.title('Training and validation acc')
plt.axis([0,12,0.7,1.0])
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```

```python
from sklearn import datasets
iris=datasets.load_iris()
X=iris.data
y=iris.target
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
```

```python
from keras.utils.np_utils import to_categorical
y_train=to_categorical(y_train)
y_test=to_categorical(y_test)
```

```python
from tensorflow.keras import models
from tensorflow.keras import layers
model=models.Sequential()
```

```python
model.add(layers.Dense(2,activation='relu', input_shape=(4,)))
model.add(layers.Dense(16,activation='relu'))
model.add(layers.Dense(32,activation='relu', input_shape=(4,)))
model.add(layers.Dense(3,activation='softmax'))
```

```python
model.compile(optimizer='rmsprop',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
history=model.fit(X_train,y_train,validation_split=0.2,epochs=100)
```

```python
import matplotlib.pyplot as plt
history_dict=history.history
loss=history_dict['loss']
val_loss=history_dict['val_loss']
epochs=range(1,len(loss)+1)
plt.plot(epochs,loss,'bo',label='Training loss')
plt.plot(epochs,val_loss,'r',label='validation loss')
plt.title('Training and validation loss')
plt.axis([0,12,0.0,3.0])
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```
